#Tokenized

```python
from pythainlp import word_tokenize
def tokenization_attachend(th_string):
  temp = ["<start>"]+word_tokenize(th_string)+["<end>"]
  try:
    temp.remove('')
  except:
    pass
  return temp
```

```python
df['tokenized_caption_th'] = df.apply(lambda x: tokenization_attachend(x['caption']), axis=1)
```
